```python
##If the information table is not present then rerun whole the beautiful soupe code
#Web scrapping for downloading all the forms , Do not run this everytime 
from bs4 import BeautifulSoup
import requests
pdf_save_path = 'C:\\Users\\Rideema Malji\\OneDrive\\Desktop\\Others\\Upwork\\CFA\\Training dataset\\Forms\\'
base_url='https://services.cookcountyclerkofcourt.org/forms/'
header = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36',
                    'referer': base_url}
html = requests.get(url=base_url,headers=header)
bs=BeautifulSoup(html.content, 'lxml')
table_body=bs.find('table')
rows = table_body.find_all('tr')
information_table = []
for row in rows:
    cols=row.find_all('td')
    ref = row.find_all('a')
    try:
        href_link_form = ref[0].get("href")
        href_link_form = base_url+href_link_form
        cols=[x.text.strip() for x in cols]
        name_form =cols[1]
        #downloading the forms 
        response = requests.get(href_link_form)
        pdf = open(pdf_save_path+str(name_form)+".pdf", 'wb')
        pdf.write(response.content)
        pdf.close()
        information_table.append(cols)
        print("File ",str(name_form), " downloaded")
    except:
        print('Error while downloading the form',cols)
```

```python
#Getting all the forms in pdf format 
import sys
from pdfminer.pdfparser import PDFParser
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdftypes import resolve1

filename = sys.argv[1]

def get_all_fillers(filenames):
    #filenames = 'C:\\Users\\Rideema Malji\\OneDrive\\Desktop\\Others\\Upwork\\CFA\\Training dataset\\Forms\\'
    form_list = os.listdir(filenames)
    form_name = []
    field_name = []
    field_value = []

    for filename in form_list:
        try:
            fp = open(filenames+filename, 'rb')
            parser = PDFParser(fp)
            doc = PDFDocument(parser)
            fields = resolve1(doc.catalog['AcroForm'])['Fields']
            for i in fields:
                field = resolve1(i)
                name, value = field.get('T'), field.get('V')
                form_name.append(filename)
                field_name.append(name)
                field_value.append(value)
        except Exception as e:
            pass
            #print('Error while processing file :',filename,e)
    field_name_decode = []
    for i in field_name:
        field_name_decode.append(i.decode('utf8', errors='ignore'))
    field_df = pd.DataFrame({'form_name':form_name,'field_name':field_name_decode})
    return(field_df)


#total_forms = get_all_fillers('C:\\Users\\Rideema Malji\\OneDrive\\Desktop\\Others\\Upwork\\CFA\\Training dataset\\Forms\\')
divorce_forms = get_all_fillers('C:\\Users\\Rideema Malji\\OneDrive\\Desktop\\Others\\Upwork\\CFA\\Training dataset\\Forms\\DIVORCE_FORMS\\')
```

```python
import pyaudio
import wave

CHUNK = 1024
FORMAT = pyaudio.paInt16
CHANNELS = 2
RATE = 44100
RECORD_SECONDS = 20
WAVE_OUTPUT_FILENAME = r"C:\Users\Rideema Malji\OneDrive\Desktop\Others\Upwork\CFA\Training dataset\Forms\DIVORCE_FORMS\sample_audio_recordings\output1.wav"

p = pyaudio.PyAudio()

stream = p.open(format=FORMAT,
                channels=CHANNELS,
                rate=RATE,
                input=True,
                frames_per_buffer=CHUNK)
frames = []

print('recording starte')
for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
    data = stream.read(CHUNK)
    frames.append(data)
    
stream.stop_stream()
stream.close()
p.terminate()

print('recording ended')
wf = wave.open(WAVE_OUTPUT_FILENAME, 'wb')
wf.setnchannels(CHANNELS)
wf.setsampwidth(p.get_sample_size(FORMAT))
wf.setframerate(RATE)
wf.writeframes(b''.join(frames))
wf.close()
```

```python
from langchain.output_parsers import StructuredOutputParser, ResponseSchema
from langchain.prompts import PromptTemplate
from langchain.chains.summarize import load_summarize_chain
from langchain import OpenAI
from langchain.output_parsers import OutputFixingParser
from langchain.chains import LLMChain
from langchain.chat_models import ChatOpenAI
transcript = "My name is Anderson ,Im a Physician Im 36 and my wife's name is julie she is 32 , we are not happy together and we have been living in IL since 2014, we want to get seperated \nwe have each others consent on this decision , we have  been married since past 8 years and have one child , we live on the street name of ABC and have cummulative income of $800k , we dont have any liability as such and the home we stay in right now is bought by both of us after marriage"
text_splitter = CharacterTextSplitter(        
    separator = "\n\n",
    chunk_size = 1000,
    chunk_overlap  = 200,
    length_function = len
)
texts = text_splitter.create_documents([transcript])

def parse_inoformation(texts,list_params):
    resp = []
    for entity in list_params[:20]:
        if 'Checkbox' in entity:    
            dtype = 'string'
            description_to_fetch = 'we have to put Either True or False here as an answer , if not sure please put False'
        else:
            dtype = 'list'
            description_to_fetch = f'From the given text what is {entity}'
        resp.append(ResponseSchema(name = f'{entity}', description = description_to_fetch,type=dtype))
    
    output_parser = StructuredOutputParser.from_response_schemas(resp)
    format_instructions = output_parser.get_format_instructions()
    
    prompt_template = """//You are a form filling assistant who picks correct information from the given page content, 
    this page content is related to the sentence given by your client based on which you have to fill the correct values, 
    Read the following text carefully and generate the values from it for filling accurate details in the format below , If you dont know the answer then please fill "NK"
    Text:
    {text[0]}
    /n{format_instructions}"""
    PROMPT = PromptTemplate(template=prompt_template, input_variables=["text"],partial_variables={"format_instructions": format_instructions})
    llm =ChatOpenAI(model = 'gpt-4',temperature=0.1,verbose=True)
    chain =  LLMChain(llm=llm,prompt=PROMPT,verbose=True)
    output=chain.run(texts)
    print(output)
    return (output_parser.parse(output.strip()))
```
